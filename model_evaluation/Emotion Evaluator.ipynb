## Calculating Current Accuracy

```python
def list_compare(predicted, actual):
    return sum(1 for a, b in zip(predicted, actual) if a == b)
```

```python
def list_set(predictions, sentiment_predictions):

    sent_influences = []
    for sentiment_prediction in sentiment_predictions:
        sent_influence = sentiment_prediction - 0.5
        sent_influences.append(sent_influence)
    
    set_predictions = []
    for i in range(0, len(predictions)):
        predictions[i] += sent_influences[i]
        if(predictions[i] > 0.5):
            set_predictions.append(1)
        else:
            set_predictions.append(0)

    return set_predictions
```

```python
def find_max_value(row):

    row_values = row.values

    if (any(value > 0.49 for value in row_values)):
        col_name = row.idxmax(axis=0)
    else:
        col_name = 'no_pred'
    
    col_dict = {
        'pred_joyful': 1,
        'pred_love': 2,
        'pred_anger': 3,
        'pred_sadness': 4,
        'no_pred': 0
    }
    
    return col_dict[col_name]
```

```python
go_emotion_res['pred_full'] = go_emotion_res[['pred_joyful', 'pred_love', 'pred_anger', 'pred_sadness']].apply(find_max_value, axis=1)
```
