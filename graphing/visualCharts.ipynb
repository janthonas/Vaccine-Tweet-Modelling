```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Data from https://reutersinstitute.politics.ox.ac.uk/digital-news-report/2023/dnr-executive-summary
news_websites = [32, 29, 28, 25, 23, 22]
social_media = [23, 24, 26, 26, 28, 30]
x_axis = [2018, 2019, 2020, 2021, 2022, 2023]
```

```python
plt.figure(figsize=(4, 3))
plt.plot(x_axis, news_websites, label='News Websites', color='#1c3ffd', linewidth=2)
plt.plot(x_axis, social_media, label='Social Media', color='#6b14a6', linewidth=2)
plt.xlabel('Year')
plt.ylabel('Proportion of People')
#plt.title('Primary way of getting news online', y=1.15)
#plt.suptitle('From 2018 to 2023', x=0.55, y=0.8, fontsize=10, color='gray')
plt.ylim(15, 40)
plt.legend()

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_color('black')
ax.yaxis.set_ticks_position('none') 
ax.xaxis.set_ticks_position('bottom') 

plt.grid(axis='y', color='grey', linestyle='--', linewidth=0.4)

plt.tight_layout()

plt.savefig('news_change_chart.png', bbox_inches='tight')

plt.show()
```
