```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data from https://reutersinstitute.politics.ox.ac.uk/digital-news-report/2023/dnr-executive-summary
news_websites = [32, 29, 28, 25, 23, 22]
social_media = [23, 24, 26, 26, 28, 30]
x_axis = [2018, 2019, 2020, 2021, 2022, 2023]
```

```python
plt.figure(figsize=(4, 3))
plt.plot(x_axis, news_websites, label='News Websites', color='#1c3ffd', linewidth=2)
plt.plot(x_axis, social_media, label='Social Media', color='#6b14a6', linewidth=2)
plt.xlabel('Year')
plt.ylabel('Proportion of People')
#plt.title('Primary way of getting news online', y=1.15)
#plt.suptitle('From 2018 to 2023', x=0.55, y=0.8, fontsize=10, color='gray')
plt.ylim(15, 40)
plt.legend()

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_color('black')
ax.yaxis.set_ticks_position('none') 
ax.xaxis.set_ticks_position('bottom') 

plt.grid(axis='y', color='grey', linestyle='--', linewidth=0.4)

plt.tight_layout()

plt.savefig('news_change_chart.png', bbox_inches='tight')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

year_month = ['2020-12', '2021-01', '2021-02', '2021-03', '2021-04', '2021-05', 
              '2021-06', '2021-07', '2021-08', '2021-09', '2021-10', '2021-11', 
              '2021-12', '2022-01', '2022-02']

tweet_counts_1 = [7326, 8185, 5716, 5505, 5364, 5150, 4447, 4757, 4948, 4484, 4247, 4386, 5779, 5313, 3310]
tweet_counts_2 = [2457, 2922, 1926, 1609, 1709, 1754, 1461, 1621, 1896, 1789, 1532, 1632, 2413, 2100, 1131]
tweet_counts_3 = [172, 216, 184, 167, 123, 119, 96, 127, 130, 92, 70, 64, 88, 96, 87]

colours = ['#2962ff', '#5747dc', '#6b14a6']

x = np.arange(len(year_month))
bar_width = 0.25
spacing = 0.05

plt.figure(figsize=(8, 3.5))
plt.bar(x - (bar_width + spacing), tweet_counts_1, width=bar_width, color=colours[0], label='Full')
plt.bar(x, tweet_counts_2, width=bar_width, color=colours[1], label='Medicine')
plt.bar(x + (bar_width + spacing), tweet_counts_3, width=bar_width, color=colours[2], label='Media')

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_color('black')
ax.yaxis.set_ticks_position('none')  
ax.xaxis.set_ticks_position('bottom')

plt.xticks(x, year_month, rotation=45, fontsize=9)

plt.xlabel('Month')
plt.ylabel('Number of Tweets')
plt.title('Monthly Tweet Counts for Three Datasets')
plt.legend()
plt.grid(axis='y', color='grey', linestyle='--', linewidth=0.4)
plt.tight_layout()

plt.savefig('monthly_tweet_counts.png', format='png', dpi=300)

plt.show()
```
